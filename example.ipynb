```python
from yoctograd.nn import Neuron, Layer, MLP
from yoctograd.engine import Value
```

```python
def loss(model, X, Y):

    inputs = [list(map(Value, xrow)) for xrow in X]

    out = list(map(model, inputs))

    loss = sum([(o - y)**2 for o, y in zip(out, Y)])

    return loss
```

```python
def train(model, X, Y):

    epochs = 100

    for epoch in range(epochs):
        
        total_loss = loss(model, X, Y)

        model.zero_grad()
        total_loss.backward()

        lr = 0.0003
        for p in model.parameters():
            p.data -= lr * p.grad

        print(f"{epoch} | loss: {total_loss.data}")
        
```

```python
model = MLP(2, [16, 16, 1])

X = [[1,0], [0,0], [1,1]]
Y = [0, 1, 1]

train(model, X, Y)

X_test = [[0,1]]
Y_test = [0]
```
