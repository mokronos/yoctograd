```python
from yoctograd import vis
from yoctograd.engine import Value
```

```python
x = Value(1.0, label="x")
w = Value(2.0, label="w")
b = Value(3.0, label="b")

y = w * x + b * 3 + 111 * 31

y.backward()
```

```python
d = vis.draw_graph(y)
d
```
